```python
highest_rated_price_df = dataset.groupby(by=['Location', 'Cuisine'])['Price for 2'].mean().reset_index() 
highest_rated_price_df.head() 
```

```python
chinese_rest_df = dataset.groupby(by='Location').agg({'Name of Restaurant': 'count', 'Price for 2': 'mean'}) \
    .rename(columns={'Name of Restaurant': 'COUNT OF RESTAURANTS'})  # Fixed renaming

chinese_rest_df = chinese_rest_df.sort_values('COUNT OF RESTAURANTS', ascending=False).head(25).reset_index()

chinese_rest_df.head()
```

```python
 fig = px.bar(chinese_rest_df, x='Location', y='COUNT OF RESTAURANTS', color='Price for 2', title= 'No. of Chinese Restaurant by Places')
 
fig.show(renderer='colab') 
```

```python
# Finding list of restaurants that have price less than and equal to 1/4th of the max price i.e Finding Cheap Restauran
 
aff_rest_df = dataset[['Name of Restaurant', 'Price for 2', 'Cuisine', 'Location', 'Top Dishes']] 
aff_rest_df = aff_rest_df[aff_rest_df['Price for 2'] <= 750] 
aff_rest_df.sort_values(by='Price for 2', inplace=True) 
aff_rest_df 
```

```python
 # Finding the highest rated list of restaurants 
 
highrate_rest_df = dataset[['Name of Restaurant', 'Price for 2', 'Cuisine', 'Location', 'Top Dishes','Dining Rating']] 
highrate_rest_df = highrate_rest_df[highrate_rest_df['Dining Rating'] >= 4.5] 
highrate_rest_df.sort_values(by='Price for 2', inplace=True) 
highrate_rest_df
```
